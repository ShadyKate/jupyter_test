Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
def convert(x):
    x = x.split()[0]
    if x.endswith('GB'):
        res = x.replace('GB', '') #реплэйс заменяет одно слово на другое сначала то что будем заменять потом чем заменим
    elif x.endswith('TB'):
        res = float(x.replace('TB', '')) * 1024 #нужно строку заменить на int 
        
    return int(res)
df ['Memory_Amount']  = df ['Memory'].apply(convert)
```

```python
df['Memory_Amount']
```

##### Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
df['Memory_Type'] = df ['Memory'].apply(lambda x: ' '.join(x.split()[1:]))
```

```python
df['Memory_Type']
```
